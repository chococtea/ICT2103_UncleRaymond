```python
import sqlite3
con=sqlite3.connect("2103UncleRaymond.sqlite")
cur=con.cursor()
```

**Table** **Creation** 

---



```python
#DengueCase Central
cur.execute("""CREATE TABLE DengueCases(
Case_ID  		INT  IDENTITY(1,1)  PRIMARY KEY NOT NULL,
Coordinates 		CHAR(20)  	NOT NULL, 
Area            		CHAR(20)  	NOT NULL, 
FMEL_UPD_D       	CHAR(20)  	NOT NULL, 
INC_CRC          	CHAR(20)   	NOT NULL, 
No_Of_Cases      	INT       	NOT NULL,
Case_Name        	CHAR(40)  	NOT NULL)"""

)
```

```python
#Aedes
cur.execute("""CREATE TABLE AedesMosquitoBreedingHabitats(
Cluster_ID        INT  IDENTITY(1,1)  PRIMARY KEY NOT NULL,
name 			        CHAR(20)	 NOT NULL,
description 		  CHAR(40) 	NOT NULL,
coordinates 		  CHAR(100) 	NOT NULL, 
type              CHAR(40)     NOT NULL)""")
```

```python
#Dengue cluster
cur.execute("CREATE TABLE DengueCluster("
"Cluster_ID  	INTEGER  PRIMARY KEY AUTOINCREMENT,"
"description 	CHAR(40)  NOT NULL, "
"properties  	CHAR(40) NOT NULL, "
"name 		CHAR(20) NOT NULL, "
"coordinates 	CHAR(100) NOT NULL, "
"geometry 	CHAR(40) NOT NULL )"

)
```

```python
#zika cluster
cur.execute("""CREATE TABLE ZikaCluster (
[ClusterID]  		INTEGER  PRIMARY KEY AUTOINCREMENT,
[name] 		CHAR(40) 	 NOT NULL,
[description]	           CHAR(100)	 NOT NULL,
[type]			CHAR(100)	 NOT NULL,
[coordinates]            CHAR(100)    NOT NULL)"""
)

```

```python
#account
cur.execute("""CREATE TABLE Account(
[AccountID]		INTEGER  PRIMARY KEY AUTOINCREMENT,
[username] 		CHAR(20)  	NOT NULL,
[password]  		CHAR(20)  	NOT NULL, 
[email]    		CHAR(20)	 NOT NULL, 
[phoneNo]  		CHAR(15) 	NOT NULL);"""
)
```

```python
#weekly 
cur.execute("""CREATE TABLE WeeklyInfectiousDiseaseBulletin (
[BulletinID] 		INTEGER  PRIMARY KEY AUTOINCREMENT,
[epiWeek]		CHAR(20)  	NOT NULL,
[diseaseName]	CHAR(50)  	NOT NULL,
[NumofCases]	INT		NOT NULL
);""")

```

```python
#Covid
cur.execute("CREATE TABLE CovidCase("
"Covid_ID  INTEGER  PRIMARY KEY AUTOINCREMENT,"
"as_of_date CHAR(30) NOT NULL,"
"age_groups INTEGER NULL,"
"status CHAR(30) NOT NULL,"
"value INT NOT NULL)"
)
```

```python
def print_TableName():
  rows = cur.execute("SELECT name FROM sqlite_master;")
  for i in rows:
    print("TableName:", i[0], "\n")

print_TableName()
```
